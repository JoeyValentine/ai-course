# 데이터 attribute방식으로도 접근 가능!
numpy랑 dictionary에서는 한column 접근 가능 but pandas는 복수개 가능

numpy.는 dict방식만, 객체는 attribute방식만 사용 가능  
data frame(pandas)에서는 attribute, dict방식 둘 다 가능
attribute방식 단점
1. attribute 이름에 공백/특수문자가 있으면 안 됨
2. 기존의 method와 이름이 같으면 충돌이 일어난다.

dict 단점
1. 귀찮다.

------

data-frame에서 column데이터 가져오는 방법
1. dict 방식 <code>data['시군구']</code>
2. attribute방식 <code>data.부상신고.mean()</code>
3. fancy indexing
4. .select_dtypes
