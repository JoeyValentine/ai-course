```python
m7 = folium.Map( location=[37.5665, 126.9780], tiles='Stamen Terrain') 

# 이런 도형들로 지도를 만들 수 있겠구나...! == 매우 노가다

# 정 n각형 만들 수 있다.
folium.RegularPolygonMarker(
    [37.5665, 126.9780], # location : 위도 경도
    popup='polygon1', # popup : 눌렀을 때 나타나는 것
    fill_color='#132b5e', # 색깔
    number_of_sides=3, # 삼각형
    radius=5
    ).add_to(m7)
folium.RegularPolygonMarker(
    [37.4665, 126.9780],
    popup='polygon2',
    fill_color='#45647d',
    number_of_sides=4, # 사격형
    radius=10
    ).add_to(m7)
folium.RegularPolygonMarker(
    [37.5665, 126.8780],
    popup='polygon3',
    fill_color='#769d96',
    number_of_sides=6, # 육각형
    radius=15
    ).add_to(m7)
folium.RegularPolygonMarker(
    [37.4665, 126.8780],
    popup='polygon4',
    fill_color='#769d96',
    number_of_sides=8, # 8각형
    radius=20
    ).add_to(m7)
m7
```

# 우리 목적 : 지역을 누르면 통계값 나왔으면 좋겠다...!

```python
import pandas as pd
import random

#Iterable
list_data = [10, 20, 30, 20, 15, 30, 45]

#Dicts of iterables
cat_1 = ['y1', 'y2', 'y3', 'y4']
index_1 = range(0, 21, 1)
multi_iter1 = {'index': index_1}
for cat in cat_1:
    multi_iter1[cat] = [random.randint(10, 100) for x in index_1]

cat_2 = ['y' + str(x) for x in range(0, 10, 1)]
index_2 = range(1, 21, 1)
multi_iter2 = {'index': index_2}
for cat in cat_2:
    multi_iter2[cat] = [random.randint(10, 100) for x in index_2]
```

```python
!pip install vincent
# vincent : 요즘은 더 이상 사용하지 않음 그러나 호환성이 좋고 만들기 쉽다. : popup에 그래프 넣기 쉽다.
# pdvega단점 : 호환성이 좋지 않다. : pdvega로 오픈하면 안되는 경우가 있다 : ex) pdvega에서 만든 그래프에서 jason파일을
# 가져와서 읽으려고 하면 지도에 안 그려지는 경우가 있다.
```

```python
import vincent
bar = vincent.Bar(multi_iter1['y1'])
bar.axis_titles(x='Index', y='Value')
bar.to_json('asset/vis1.json')
```

```python
line = vincent.Line(multi_iter1, iter_idx='index')
line.axis_titles(x='Index', y='Value')
line.legend(title='Categories')
line.to_json('asset/vis2.json')
```

```python
scatter = vincent.Scatter(multi_iter2, iter_idx='index')
scatter.axis_titles(x='Index', y='Data Value')
scatter.legend(title='Categories')
scatter.colors(brew='Set3')
scatter.to_json('asset/vis3.json')
```

```python
area = vincent.Area(list_data)
area.to_json('asset/vis4.json')
```
