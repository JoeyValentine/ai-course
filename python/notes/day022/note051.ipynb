# 전처리 할 때, for는 우아하지 않으므로 map을 사용하도록 하자.

```python
import numpy as np
```

```python
tips.pivot_table('total_bill', ['sex', 'day'], 'smoker', aggfunc=np.sum, margins=True)
# 엑셀처럼 사용할 수 있음 , groupby보다 사용하기 쉽다/직관적임.
# groupby를 엑셀처럼 사용할 수 있다.
# values : 표 안에 값
# 가로 : column
# 세로 : index : ['sex', 'day'] : multi index 기법
# aggfunc : 어떤 함수를 적용시킬까 : default로 평균을 구해줌
# margins : 부분합 보여줌 : 회계사들이 가장자리에서 계산해서 margin이라고 함
```

# Multi indexing

```python
import pandas as pd
```

```python
pd.crosstab([tips.sex, tips.day], tips.smoker).index
# levels : multi index의 각각의 값
# labels == codes == [0, 0, 0, 0, 1, 1, 1, 1] : 문자를 숫자로 바꾼 것을 label이라고 함 : enum과 유사 : 남자는 0, 여자는 1ㅜ
```

```python
pd.crosstab([tips.sex, tips.day], tips.smoker).index.levels.append('sss')
```

```python
pd.crosstab([tips.sex, tips.day], tips.smoker).index.levels
# immutable인데,
```
