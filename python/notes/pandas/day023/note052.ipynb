```python
# axes는 한 figure에 그릴 수 있다.  
# plt.figure를 안 사용하면 plt.axes에서 자동으로 figure생성
plt.axes([0,0,1,1]) # : figure에서 상대적인 거리 개념
plt.axes([0.3,0.3,0.5,0.5])
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
plt.style.use('ggplot')
# 하나하나 설정하기 귀찮아서 style사용한다.

# seaborn pandas도 matplotlib으로 바꿀 수 있다. 왜냐하면 matplotlib기반으로 만들었기 때문이다.
```

```python
tips.groupby('sex')['tip'].mean().plot.bar(label='a', figsize=(12,4))
plt.legend()
# plot.bar는 
# matplotlib 함수에 label='a', figsize=(12,4)를 넘겨줌
```

```python
tips.groupby('sex')['tip'].mean().plot.bar(label='a', figsize=(6,6), ylim=(2.7,3.3))
plt.legend()
# 사기치기...
```

```python
plt.plot([3,4,5], label='lggg') # 라벨이 지정 되어야
plt.legend(loc=1) # 이걸 호출할 때 legend를 표시해줌, loc == 위치
```
