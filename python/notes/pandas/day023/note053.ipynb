```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
plt.style.use('ggplot') # 개간지
sns.pairplot(iris, hue='species', diag_kind='kde')
# seaborn은 pandas기반임
# hue : 색깔별로 groupby해줌
```

```python
# 인공 데이터 만들기
from sklearn.datasets import make_blobs
```

```python
data = make_blobs()
```

```python
data_feature = pd.DataFrame(data[0], columns=['x', 'y'])
# 데이터 프레임 인스턴스 방식으로 만듦
```

```python
data_feature
```

```python
data_target = pd.DataFrame(data[1], columns=['target'])
```

```python
data_target
```

```python
data = pd.concat([data_feature, data_target], axis=1)
```

```python
data
```

```python
data.info()
```

```python
data.target.value_counts()
# 0, 1, 2는 원래 category구나..!
```

```python
sns.pairplot(data)
# 0, 1, 2를 숫자로 인식해서 그래프를 그림... category로 수정해 줘야 함
```

```python
sns.pairplot(data, hue='target')
```

```python
data.target = data.target.astype('category')
```

```python
data.info()
```

```python
sns.pairplot(data, hue='target')
# type이 바뀐 줄 알지만 사실 아님 ㅂㄷㅂㄷ

# category에서 지원하는 기능을 쓸 수 있을 뿐, 실제 값은 안 바뀐다. 즉, 숫자처럼 사용된다. (int와 category를 다중상속이라 생각해도 좋다.)
# int(target)에서 category기능만 추가됨... 

# 카테고리로 바꿨는데도 에러...??>?? pairplot 그래프를 그릴 때 같이 숫자로 구분하려고 해서 오류가 남

# 데이터가 숫자처럼 생겼지만 category기능을 지원함
```

```python
sns.pairplot(data[data.columns[ : -1]], hue='target')
# fancy indexing과 유사
# data[data.columns[ : -1]] : 얘는 target column이 없음
```
