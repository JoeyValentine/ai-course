# 데이터 유지보수(계속 유입되는 정보 관리하려면) 하려면 data format을 잘 설계해야 한다.

pandas는 예측분석 못함,,, 
1. 정형 데이터 불러옴 : pd.read_ => rename, long format으로 변환해준다.
2. 필요한 정보를 indexing, slicing으로 통계값을 구한다.  
info, describe, tail, head, sample를 확인함 => 예측분석 가능한가 결정함 == 데이터가 숫자로 이루어 졌냐, 분리시킬 수 있나?   
데이터 품질 확인 : boxplot(), sns.pairplot() : 일반적인 자연 현상은 정규분포를 따른다. 따라서 정규분포인지 확인한다. 정규분포면 잘 뽑은거다.    
corr, cov, skew,  

원하는 데이터를 뽑아내기(slice, indexing) : 뽑아내야 원하는 통계값 구함 & 잘못된 것을 수정함의 관점...  
3. 예측 가능하면 예측 가능한 형태로 변환(encoding)
4. 성능이 나쁘면 다른 기법을 사용, 그래도 안되면 기초 통계분석 사용함 (예측의 과정은 padas가 못한다.)

예측 가능성 확인하는 방법  
*숫자들이 class target label를 대표할(예측할) 경우  
*각 원소들이 논리적 상관관계가 있는 경우 (효율은 고려 안하고...)

예측 못하면 기초통계분석 한다. value_counts, unique, nunique, aggregation 분석(groupby series) 사용
