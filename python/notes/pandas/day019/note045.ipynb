# [black box](https://dreamgonfly.github.io/2017/11/05/LIME.html)
위의 과정은 깊이있는 통계분석 하기보다는 데이터의 품질을 확인하기 위해서 위의 과정을 수행함  
공학적 관점으로 접근함 : 통계적 설명가능성/타당성을 증명하기보다는, 속도를 더 중시함  
mri분석 같은거 설명 안하고 예측함(정확성이 더 높음..) : but 만든 사람(black box)도 이해 못함, 설명 불가능 : 사고나면 누가 책임지나...   
법적 문제로 사용하지 못한다.  
따라서 블랙박스 모델(설명 안되는 모델) 말고, 설명 가능한 모델을 만들어야 한다.  
# [블랙박스 AI 아닌 '설명 가능 인공지능'이어야 하는 이유](http://www.ciokorea.com/news/38958)

```python
data.corr()
# 절댓값이 1에 가까울 수록 관계가 크다.
# 음수면 : 내가 커질수록 남이 작아진다.

# 사용하는 이유
# 1. 상관관계가 크면, 데이터를 없애도 된다. == 오캄의 면도칼 == 데이터를 없앨수록 좋다.
# 2. 다중공산성을 확인하기 위해서
```

```python
data.cov()
# 사용하는 이유
```

```python
data.skew()
# 치우쳐진 정도 확인
# +면 왼쪽으로, -면 오른쪽으로 치우쳐짐
```

# 예측 모델로 만들기 위해서 string값을 숫자로 변환시켜줌 (encoding)

예측 품질은 보장 못한다...

```python
tips = sns.load_dataset('tips')
```

```python
tips.sex = tips.sex.map({'Female':0,'Male':1})
# 여자를 0으로 대응, 남자를 1로 대응
```

```python
tips
```

```python
tips.smoker = tips.smoker.map(lambda x : 0 if x=='N' else 1)
# 이러한 과정을 전처리라고 한다.
```
