```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris.info()
# info로 확인해야 할 것
# column 수 확인해야 함 : 차원의 저주 : 모델이 간단할수록 좋다.(column이 적을수록 좋다.) == 분석하기 좋다.
# data 수
# data type : 
# category는 범위가 정해져 있음
# pandas가 빠른 이유 : 모든 데이터를 메모리에 올리기 때문이다. (memory usage: 5.9+ KB)
```

```python
iris.describe()
# 숫자는 보기 힘드니까 seaborn에 있는 pairplot, boxplot,  을 확인하자. (데이터를 이해하기 위해서)
```

```python
sns.pairplot(iris, hue='species') # 팔레트 키워드 옵션으로 색깔 바꿀 수 있다.
# hue : column명으로 색깔을 구분하고 싶다. 라는 의미

# 그림으로부터 알 수 있는 것
# 데이터가 분리 가능한가
# 데이터의 분포 상태
# 데이터가 종류마다 어떻게 분포가 되어있는가.

# 정규 분포에 대한 정리
# 꽃의 크기는 자연 현상이므로 정규분포를 따른다.
# 중심 극한 정리 : 데이터가 많을수록 정규분포를 따른다.
```

```python
iris.boxplot()
# 데이터 분포 확인
```

```python
iris.corr()
# 상관 관계값의 절댓값이 크면 상관관계가 크다. 
# 양수면 같이 커지는 경향, 음수면 서로 반대로 커지고 작아짐

# 상관관계가 크면, 둘 중 하나 없애도 된다. == 차원의 저주 

# 인과관계로 해석해서는 안된다. 같은 경향이 있다고 파악하는게 옳다.
# 잎이 커서 sepal이 크다 nono...

# 숫자는 보기 힘드니까 sns.heatmap(그림)을 사용하자
```

```python
sns.heatmap(iris.corr(), annot=True)
# 연관관계가 클수록 1에 가깝다.
# 색깔로 연관관계 표현
```
