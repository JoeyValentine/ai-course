````python
def a(t):
    return t + 1
````

````python
a = lambda t : t + 1
# 람다함수식
# 콜론 왼쪽은 parameter, 왼쪽은 return값
# 이름이 없어서 익명함수라고도 한다.
# 식이라서 할당 가능 
````

````python
a(4)
# 관례상, python에서는 a를 name또는 식별자라 한다. 함수도 값으로 들어올 수 있기 때문이다.
# 다른 언어에서는 변수라고 함
# 값이라고 해도 틀린말은 아님. first class function
````

# lambda함수 용도 : 재사용할 필요없는 함수 만들 때 씀

map에서 한 번 쓰기 위해서 함수를 따로 def하기에는 너무 귀찮아.... lambda로 편리하게 해결하는게...! 

재사용하는 함수일 땐 함수 정의해서 쓴다.

````python
def z(x):
    return x + 1
map(z, [1, 2, 3, 4])
````

````python
map(lambda x : x + 1, [1, 2, 3, 4])
````

``` python
def make_incrementor(n):
    return lambda x: x + n
f = make_incrementor(42)
f(0)
f(1)
```
lambda 함수로 runtime때 closure생성 가능하다!
