```python
import matplotlib.pyplot as plt
```

[parameter](https://legacy.python.org/dev/peps/pep-3102/)

```python
plt.plot()
#  가변인자, default에 의해서 아무것도 입력 안해도 됨
#  plot([x], y, [fmt], data=None, **kwargs)
# []는 optional을 의미한다.

# plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
# scalex, scaley, data는 keyword 방식 아니면 바꿀 수 없다. positional은 args가 다 흡수해서...
```

```python
plt.plot([3, 4, 80])
# y의 값이 2, 4, 6으로 정해짐
```

```python
# 가변 인자로 overloading 안됨을 극복함
plt.plot([1, 2, 3], [4, 5, 6])
# x값에 1,2,3
# y값에 4,5,6
```

```python
plt.hist([1,2,1,1], color='pink')
# 뭘 몇개를 입력할지 말지 모르겠는걸??? x의 값이 안 정해져 있으므로 1개만 입력해도 괜찮음
# keyword방식 장점 : 바꾸고 싶은것만 바꿀 수 있음 : color처럼 keyword 방식으로 해도된다.
```

```python
plt.bar
# x하고 height 값이 없으므로 2개 필수적으로 써야함
# *뒤에는 무조건 keyword 방식
```

```python
plt.bar([1,1,3,2], 1, 0.2)
# plt.bar([1,1,3,2], 1, width = 0.2) positional도 되고 keyword도 되고
```

```python
plt.bar([1,1,3,2], 1)
```

```python
# statement 종류 5개
# 할당문 : 6가지
# 조건문 복잡하게 하는 것 : elif , and/or, 중첩if
# 반복문 : 2가지 : for while
# 선언문 : 함수, 클래스 선언 2가지

# else 용례
# if, 반복문, 예외처리

# 함수 사용하는 법 5가지
# positional 
# keyword : 순서 바꿔도 됨 , keyword 한번 쓰면 끝까지 keyword 써야 함
# positional only : \
# keyword only : 맨 앞 *

# python에는 switch case가 없음 
# 할당 방법 6 가지
# a = 1 : built in 함수를 덮어쓰지 않도록 주의!
# a = b = c = 1 : mutable 할당 할 때 주의
# a += 1 : ++a, a++ 파이썬에는 불가능

# a, b = 2, 3 : 나눠서 대입하면 안됨 나눌꺼면 temp 변수 따로 써!
# a = b
# b = a 불법!

# a, *b *들어간 할당 나머지
# global a : 글로벌 할당 가능

# *a : 나머지 할당 : type은 list가 됨
# 단독 별표 '*' : 뒤에는 모두 keyword
# *a : 가변 positonal
# **a : 가변 키워드
```

```python
a, b = 3, 2
a, b = b, a # python에서 자동으로 temp를 만들어서 바꿔줌
print(a, b)
```
