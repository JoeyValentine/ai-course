# clustering

```python
from sklearn.cluster import KMeans
# knn과 유사
```

```python
km = KMeans(3)
# 데이터 형태를 모르기 때문에 cluster가 몇 개 인지 모른다.
```

```python
km.fit(iris_data)
```

```python
vars(km)
# labels_ 가 핵심

# 객체가 가지는 attribute확인
# vars vs dir ????차이가 뭐더라
```

# 원래 데이터에는 target이 없는데(없다고 가정), 가상으로 target을 만듦

원래 데이터에는 target이 없음  
clustering으로 target data생성 == supervised learning에 활용 가능해짐

```python
iris_data['target'] = km.labels_
```

```python
iris_data
```

# 정규화 해주는 이유 == 값이 크면 clustring에 영향을 미치기 때문이다.
값이 너무 커서, outlier를 한 덩어리로 묶어버리는 경우가 생긴다.

> Clustering outliers.
> 
> Centroids can be dragged by outliers, or outliers might get their own cluster instead of being ignored.  
> Consider removing or clipping outliers before clustering.

```python
iris_data = iris_data.drop(columns='target')
```
