# [pipeline caching](https://scikit-learn.org/stable/modules/compose.html#caching-transformers-avoid-repeated-computation)

```python
from sklearn.pipeline import Pipeline, make_pipeline
# 여러개의 일을 연속으로 동시에 일을 해줌, 귀찮게 다 안쳐도 된다.
# 단점, 중간단계 확인이 불편하다.
```

```python
pipe = Pipeline([('sc', StandardScaler()), ('knn', KNeighborsClassifier())], 'cache')
# StandardScaler한 이후에 KNN해라...!
# (이름, 함수) : 이름쓰기 귀찮아... make_pipeline쓰기...!
# fit할 때 cache생성됨

# 오류가 생겨서 중단되면, cache가 없으면 처음부터 다시 해야 함 ㅂㄷㅂㄷ
# 만일 cache를 만들면, 중단된 시점부터 다시 시작한다.
```

```python
pipe.fit(wine_dataframe.iloc[:,:-1], wine_dataframe.iloc[:,-1])
# 작업 directory에 cache가 생김 : 시간이 오래걸리는 작업에서 유용함. 중간에 멈추는 경우 유용함
# 단계별로 임시파일저장함. 단계별로 임시저장해서 중간에 멈추는 경우, 임시데이터부터 다시 시작...
```

scikit.preprocessing, scikit.pca만 fit_transform해야 함(모양을 바꾸고자 함)
나머지는 모두 fit

```python
vars(pipe)
# steps에서 이름은 별로 안 중요하다. == 쓰기 귀찮다.
```

```python
pipe = make_pipeline(PCA(4), StandardScaler(), KNeighborsClassifier())
# 이름 안 써도 됨, 더 편해..!

# predict할 때, 추측할 데이터 scaling안해도 되는건가?
pipe.predict(wine_dataframe.iloc[0,:])
```
