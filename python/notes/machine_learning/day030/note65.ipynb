```python
# 알고리즘과 hyper parameter 한번에 찾기
# pipeline을 통해서만 가능함..!
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
gb = GaussianNB()
kn = KNeighborsClassifier()
```

```python
from sklearn.base import BaseEstimator
class Dummy(BaseEstimator):
    def fit():
        pass
    def score():
        pass
```

# 공식적으로 scikit에서 지원하는게 아님... ㄸㄹㄹ

```python
# 핵심
# 알고리즘 이름과 hyper parameter이름 맞춰줘야 한다.
param_grid = [
    {
        'clf' : [gb]
    },
    {
        'clf' : [kn],
        'clf__n_neighbors' : [2,3,4,5,6,7] 
        # 왜 __ 2개 여야함???
        # _ : 한개는 snake방식을 나타냄
        # __ : 관례상 underbar 2개는 접두사를 나타냄, under-bar 2개 써야한다.
        # __ 2개를 써서 일어날 지 모르는 이름충돌을 막는다. == name mangling
    }
]
```

```python
# 재학습을 못함... ㄸㄹㄹ 
```

```python
grid = GridSearchCV(pipe, param_grid, cv=10)
# GridSearchCV는 pipe라인과 연동해서 쓸 수 있다.
# 조건 : 파이프라인에 적어도 하나 estimator가 있어야 한다.
```

```python
grid.fit(iris.iloc[:,:-1], iris.iloc[:,-1])
```

```python
grid.cv_results_
# 보기 안좋아서 pandas data frame써서 보자..!
```

```python
import pandas as pd
pd.DataFrame(grid.cv_results_).T
# pipe라인과 GridSearchCV를 같이 사용해서
# 알고리즘별로 성능을 구별함
```

```python
grid.best_params_
# 최고의 성능을 가지는 것을 알려줌
```
