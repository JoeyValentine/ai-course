# 쪼갰는데 데이터가 작으면 어떻게 해야할까???
[data leakage](http://blog.naver.com/PostView.nhn?blogId=tjdudwo93&logNo=221085844907&redirect=Dlog&widgetTypeCall=true&directAccess=false) == 데이터가 작을 때 자주 발생함  
가능한 모든 데이터(경우의 수)를 학습시켜야 한다.

trainning set에는 없는데 test set에 있는 데이터가 있으면 성능이 떨어진다.

# 해결책 == [cross validation](https://3months.tistory.com/321)

데이터 집합과 학습 집합을 모든 경우의 수로 나누어서 정확도를 평균시킨다.  
테스트데이터, 학습데이터  
1, {2,3,4,5}  
2, {1,3,4,5}  
3, {1,2,4,5}  
...

```python
from sklearn.model_selection import cross_val_score
```

```python
t = KNeighborsClassifier() # hyper parameter & algorithm에 따라서 성능이 바뀐다.
# t = KNeighborsClassifier(4) : 4 : hyper parameter
cross_val_score(t, iris.iloc[:,:-1], iris_pd.iloc[:,-2] , cv=10).mean() # 여기엔 쪼갠 데이터 말고 모든 데이터 넣어야 함
# cv == 등분
# estimator : 알고리즘의 인스턴스

# 데이터가 많으면 cross validation이나 아니냐 상관없이 정확도가 비슷하다. == 그냥 안쓰는게 나음(속도가 느려짐)
# 데이터의 수가 적은 경우 cross validation이 낫다.
```
