# [pivot vs pivot_table](https://nikgrozev.com/2015/07/01/reshaping-in-pandas-pivot-pivot-table-stack-and-unstack-explained-with-pictures/)

```python
data.pivot('user_id', 'movie_id', 'ratings').fillna(0.)
# pivot은 index column만들어줌
# pivot table은 여러 개 더하거나 곱할 때 사용하는 거

# pivot과 pivot table차이가 뭐지? : 중복된 entries에 대해서 연산을 수행하는 것 같다.
```

```python
# 여태 상관관계구할 때 column 원소들끼리에 대한 것이다.
matrix = data.pivot('user_id', 'movie_id', 'ratings').fillna(0.)
```

```python
matrix_corr = matrix.T.corr()
# 유사도 matrix
```

```python
matrix_corr[3].sort_values(ascending=False)[1:4]
# 사용자 3번과 가장 가까운 3명의 사람들
```

```python
idx = matrix_corr[3].sort_values(ascending=False)[1:4].index
```

```python
my_movie = data[data.user_id==3].movie_id
# 3번이 본 영화 ID
```

```python
other_movie = data[data.user_id.isin(idx)].movie_id
# 863, 616, 489의 결과를 다 합하고 싶은데...!
# 3명들이 본 영화들
```

```python
recommend = set(other_movie) - set(my_movie)
# 나한테 추천하는 영화 ID
# 다른 사람이 본 영화

# 집합 연산을 더 멋지게 할 수 있을까???
```

```python
recommend_id = data[data.movie_id.isin(recommend)].movie_id
# 내가 안 본 영화들
```

```python
from surprise import Dataset
```

```python
dir(Dataset)
# 'load_from_df' : data frame으로부터 데이터를 읽어오기
```

```python
data
```

```python
from surprise import Reader
from surprise import SVD
```

```python
s_data = Dataset.load_from_df(data, Reader(rating_scale=(1,5)))
# 공식문서가 틀렸음
# pandas힘들게 안써도 됨, 데이터만 불러와도 됨 & 한줄로 해결 대박
# surprise로 한방에 해결
```

```python
s_data
```

```python
from surprise import KNNBasic
```

```python
knn = KNNBasic()
# sim-option에서 False하면 아이템끼리 유사도 측정 but contents based가 아님
```

```python
dir(s_data)
```

```python
knn.fit(s_data.build_full_trainset())
```

```python
knn.predict(3, 758)
# 3번에게 758 추천해줬더니 est==1.7점 나옴

# 영화를 추천했을 때 몇점이(점수만) 나올까만 예측할 수 있다. == 다양한 알고리즘을 사용할 수 없다.
# 다양한 알고리즘을 적용하려면 pandas로 직접 model을 만들어야 한다,.
```

```python
for i in set(recommend):
    print(svd.predict(3,i))
# 추천 알고리즘은 검증할 방법이 없어서 사람으로 검증함(정답이 없기 떄문이다.)
# 보통 predict는 train-data가 정답인 것들로 가정하기 때문이다.
# 3 * 3 == 3 * 2 * 2 * 2 * 2 * 3 인 행렬로 쪼개서 
```
