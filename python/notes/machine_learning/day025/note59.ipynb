```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
# 가정 : label이 정확하다고 가정
# 일부는 학습시키고 일부는 확인하는 데이터로 나눈다.
# 100개로 학습하고 50개로 확인한다.
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris.species = iris.species.map({'setosa':0,'versicolor':1,'virginica':2})
# 학습하기 위해서 데이터를 수정함
```

```python
iris
```

```python
# hold out 방식
X_train, X_test, y_train, y_test = train_test_split(iris[iris.columns[:-1]], iris.species)
# 반복할 때마다 랜덤으로 쪼갠다.

# X_train, X_test, y_train, y_test = train_test_split(iris[iris. columns[:-1]], iris.iloc[:,-1])
# fancy indexing, iloc

# x값 y값을 random하게 비율로 잘라준다.

# X 쪼개고, Y쪼개고...
```

```python
from sklearn.neighbors import KNeighborsClassifier # 0 1 2 로 분류할꺼라서 KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

# learning(CS) & fitting(통계학 용어) 같은 말이다.

```python
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test) == y_test.values
# 각 데이터를 예측

# 데이터가 작아서 일반화된 모델이 안 나올 경우가 있다.
# 랜덤하게 잘라서 결과가 매번 다르다.

# 데이터에 따라서 성능이 변하면 학습이 덜 된 것이거나 잘못된 데이터...
```

알고리즘보다 데이터가 더 중요하다 == The Unreasonable Effectiveness of Data

# 데이터를 엄청나게 확보하면 성능을 보장할 수 있음을 나타냄

```python
편차 : 흩어진 정도  
bias : 편향 : 
        
목표 : 편향도 작고 variance작은 모델을 만드는 것이 목표
```

bias & variance tradeoff == 최적 == 둘 다 적당히 작은 값  
var bias 두 개 다 작아질 수 없다.  
상수(에러) = 상수^2 + 편차 + 편향^2  
편향이 작아지면 편차가 커지고  
편차가 작아지면 편향이 커진다.

일반적인 최적의 모델 찾는것이 목적(미지의 데이터에 대해서도 성능이 좋아야 한다.)  
모델이 복잡하면 복잡해질수록 overfitting 문제가 일어남(실제 성능이 안좋다)  
== 오캄의 면도칼, 차원의 저주

```python
underfit == 데이터가 적어서 학습이 잘 안되는 경우
```

best generalization 찾는 것이 최적의 모델을 찾는 것이다 == 어렵다..  
학습을 시킬수록 training error는 줄지만(bias) variance가 커진다.  
즉, bias가 너무 낮다고 좋은 모델이 아니다.

```python
knn.predict(X_train) == y_train.values
# 우아하지 못함 ㄸㄹㄹ...,
```

```python
from sklearn.metrics import confusion_matrix # 함수,,?
from sklearn.metrics import classification_report # 함수,,?
```

```python
confusion_matrix(y_test, knn.predict(X_test))
# 행이 실젯값인지 열이 실젯값인지 확인해봐야 한다.

# 행이 총 갯수, 열이 맞춘 갯수
# 실제 12개인데 12개 맞춤
# 실제 12개인데 11개 맞춤
# 실제 14개인데 14개 맞춤

# 상황에 따라서 1종오류, 2종오류 심각성이 다르다.
```

```python
No Free Lunch(반드시 댓가를 치룬다.)
1.  어떤 것은 가정을 해야한다. 가정이 현실세계에 안 맞는다. == 댓가를 치룬다. == 특정 상황에서만 옳다.
일반화된 

# 그나마 transfer learning은 덜하다... 
```

```python
classification_report(y_test, knn.predict(X_test))
# \n있으면 print해주자...
```
