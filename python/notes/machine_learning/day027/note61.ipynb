# groupby 분석 3가지

```python
titanic.groupby(['sex']).survived.sum()
# 합을 구하면 생존자 수를 구할 수 있다.
# pivot table
# cross table
```

```python
titanic.pivot_table('survived', 'sex')
# 결론 : 여자가 더 많이 살아남았다.
```

```python
titanic.groupby(['sex','class']).sum()[['survived']]
```

```python
titanic.groupby(['sex','class']).mean()[['survived']]
# 노 젓는 사람들이 많았다... == Third class 남자들 노 저으라고 시킨듯..?
```

```python
titanic
```

```python
titanic[titanic.age.isnull()].survived.value_counts()
# 52명은 살았지만 나이를 안적음...
```

```python
titanic[titanic.age.isnull() & titanic.survived==1]['class'].value_counts()
```

```python
age_index = titanic[titanic.age.isnull()].index
# 나이대별로 알고싶어서 
# NaN 나이만 지워보자...
```

```python
age_titanic = titanic.drop(index=age_index)
# age_index에 있는 index만 지우기
```

```python
# 나이 구간대 나누는 함수 pd.cut, pd.qcut
pd.cut(age_titanic.age, 10)
# 10구간으로 알아서 쪼개줌 
# 
```

```python
pd.qcut(age_titanic.age, 10)
```

```python
# but map을 사용하자...!
def age_(x):
    return 10 * (int(x)//10)
# 몇 십대인지 확인
```

```python
age_titanic.age = age_titanic.age.map(age_)
```

```python
age_titanic
```

```python
age_titanic.groupby(['sex','age']).mean()[['survived']].plot.bar()
# 보기 안좋아서 unstack사용
```

```python
age_titanic.groupby(['sex','age']).mean()[['survived']].unstack().plot.bar()
```
