```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
import pandas as pd
```

```python
t = pd.get_dummies(iris.species)
```

```python
iris_ = pd.concat([iris, t], axis=1)
```

```python
iris_.drop(columns='species', inplace=True)
```

```python
iris_
# y에 3개 넣어도 3됨???
```

```python
# One-hot encoding으로 가능한가..?
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_.iloc[:,:-3],iris_.iloc[:,-3:])
# y가 여러개인 경우 어떻게 할까..?
```

```python
y_test
# one-hot encoding으로 y 여러개 사용 가능..!
```

# One-hot encoding사용하면 모든 알고리즘을 사용할 수 없다.
target data에서는 label encoding써도 됨   
target data에서는 숫자를 크기로 인식 안하고 category로 인식해서 학습할 때 문제가 없음  
학습데이터인 경우에는 숫자를(원래는 category인데) 크기로 인식해서 문제가 생김  
따라서 학습데이터는 one-hot encoding써야 함

```python
from sklearn.linear_model import LogisticRegression
# 주로 2종류 구분할때만 사용
```

```python
lr = LogisticRegression()
# 알고리즘이 2부분만 구분 가능하다..
# A B C 3 category분류하는 방법...!
# A/B & A/C & B/C구분하는 방식으로 3번구분해서(multi class 방식으로) 어디에 속하는지 판별

# 이런 알고리즘에 one-hot쓰면 에러날 가능성이 있다.
```

```python
lr.fit(X_train, y_train)
```

# 데이터 분포 형태 구분 방법

데이터 분포형태를 구분하면  
분포에 관련된 알고리즘들끼리 성능비교하면 된다.  
다 구할 필요가 없다.

linear regression : 용도에 따라서 regression, classifcation할 때 사용할 수 있음
logistic regression = 이름은 regression인데, 기능은 classification하는 것임  

linear algorithm : 단순함, 속도가 빠름, 설명 가능성(ex/ logistic regression)  
non-linear algorithm : 성능이 높아서 overfitting됨, 속도가 느림(ex/ naive baysian == 제일 간단한 알고리즘)  
**두 알고리즘 간단한거 하나씩만 써봐서 더 빠른 알고리즘으로부터 데이터 분포를 알 수 있다.**  
다음부턴 빠른 알고리즘으로만 작성한다. (kaggle)  
overfitting문제가 제일 어려움...(내 데이터만 맞추는 경우)
