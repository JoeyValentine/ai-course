pandas많이 쓰는 3총사  
intersection, isin, where

````python
import numpy as np
````

````python
a = np.arange(10)
````

where의 용법 2가지  

````python
np.where(a > 3)
````

````python
np.where(a > 3, 1, 0)
# 참일 때 1, 거짓일 때 0
````

numpy는 조건에 맞아도 안맞아도 바꾸는 where사용  
pandas는 조건에 맞으면 그대로, 다르면 바꿔주는 방식
