# 다중상속

```python
class A:
    def __init__(self):
        print('A')

        
class B(A):
    def __init__(self):
        print('B')
        A.__init__(self)

        
class C(A):
    def __init__(self):
        print('C')
        A.__init__(self)
        
        
class D(B, C):
    def __init__(self):
        print('D')
        D.__init__(self)
```

```python
d = D()
# 재귀는 일정 깊이 이상 못한다. 게다가 자동으로 멈춰준다.

# but 그냥 함수 무한 반복(while)은 가능! 메모리 사용 안하니까
```

## 상속의 경우 연산 순서를 바꿀 경우, 연산 순서가 바뀌면서 결과가 달라질 수 있으므로 조심해야 한다.

```python
class A:
    def __init__(self):
        print('A')

        
class B(A):
    def __init__(self):
        A.__init__(self) # A대신 super로 바꿀 수 있다.
        print('B')

        
class C(A):
    def __init__(self):
        A.__init__(self)
        print('C')

        
class D(B, C):
    def __init__(self):
        B.__init__(self)  # function 방식 : 클래스명이 먼저 앞에 온다.
        C.__init__(self)
        print('D')
        # B의 생성자를 다시 호출하므로 재사용성이 높다.
        # method 방식 : 인스턴스 뒤에 점을 찍어서 호출
        
```

```python
d = D()
```

```python
vars(D)
# method : self 를 생략해도 된다.
```

```python
class A:
    def __init__(self):
        print('A')

        
class B(A):
    def __init__(self):
        super().__init__() # super는 method 방식이다.
        print('B')

        
class C(A):
    def __init__(self):
        super().__init__()
        print('C')

        
class D(B, C):
    def __init__(self):
        super().__init__()  # function 방식 : 클래스명이 먼저 앞에 온다. && 다중상속을 때 자기가 알아서 순서대로 호출해준다.
        print('D')
        # B의 생성자를 다시 호출하므로 재사용성이 높다.
        # method 방식 : 인스턴스 뒤에 점을 찍어서 호출
        
```

Super는 method방식이다. (method는 첫번쨰 인자를 안쓴다.)
 vs
클래스명을 명시해서 __init__을 비교하는 경우

```python
d = D()
# 스택에 넣어서 꺼내면서 순서를 결정한다.
# 스택 앞 : [D, B, C, A] : stack 에 이렇게 들어감 : A -> C -> B -> D 순서로 호출함 (stack역순)

# 출력이 아까 예제와 다르다.
# super는 부모대신 부르지만, super 쓸때와 아닐 때 행동방식이 다르다. 
```

```python
D.__mro__
# method resolution order : 다중상속 실행 순서 알려줌 
# 스택 앞 : [D, B, C, A] : stack 에 이렇게 들어감
# 가득 차면 하나씩 꺼내면서 함수를 호출한다.
```
